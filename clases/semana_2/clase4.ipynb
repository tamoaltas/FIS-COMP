```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2*np.pi,2*np.pi,100)
w = [1, np.pi,15, 30]

for w_i in w:
    y = np.sinc(w_i*x)

    # Las llaves est√°n para decir d√≥nde inicia/termina el comando \rm
    plt.plot(x,y, label=r'${\rm sinc ( %2.2f x)}$' % w_i)

plt.title(r'${\rm sinc (\omega x)}$')
plt.legend()
plt.show()
```

```python
# Otra manera de correr un ciclo for
x = np.linspace(-2*np.pi,2*np.pi,100)
w = [1,0.5,0.25]

for i in range(len(w)):
    y = np.sin(w[i]*x)

    # Las llaves est√°n para decir d√≥nde inicia/termina el comando \rm
    plt.plot(x,y, label=r'${\sin ( %2.2f x)}$' % w[i])

plt.title(r'${\sin (\omega x)}$')
plt.legend()
plt.show()
```

Para hacerlo de forma modular, podr√≠a primero definirse
una func√≥n Heaviside para 1 sola x, y luego hacer que 
Heaviside_arreglos use esta √∫ltima.

```python
def Heaviside_arreglos(Xs, x0=0):
    H = np.zeros_like(Xs)
    
    for i in range(len(Xs)):
        if Xs[i] <= x0:
            H[i] = 0
        else:
            H[i] = 1

    return H

x = np.linspace(-5,5)
y = Heaviside_arreglos(x,-1)

plt.plot(x,y)
plt.show()
```
